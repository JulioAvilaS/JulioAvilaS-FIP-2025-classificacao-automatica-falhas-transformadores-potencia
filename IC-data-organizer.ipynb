## Utilizar a biblioteca pandas para verificar as linhas iguais

```python
filename = 'Datasets/Original-Datasets/Original-Datasets-csv/disordered_dataset1_12.csv'
df = pd.read_csv(filename)
samples_list = df.values.tolist()

print("Amostras originais:", samples_list)
```

```python
count = 0
duplicates = []
seen = set()

for item in samples_list:
    tuple_item = tuple(item)
    if tuple_item in seen:
        count += 1
        duplicates.append(item)
    else:
        seen.add(tuple_item)

print("\nItens duplicados:", duplicates)
unique_samples = [list(item) for item in seen]
```

## Transforma, divide as labels e armazena com os valores corretos:

```python

# labels_list = []
# df_rounded = df.round(0)

# for row in samples_list:
#    row_list = list(row)  
#    last_element = row_list.pop()  
#    labels_list.append(int(last_element) + 1)   


# print(labels_list)   
# print(samples_list)   
# print(len(samples_list))

labels_list = []
cleaned_samples = []

for row in unique_samples:
    labels_list.append(int(row[-1]) + 1)  # Pega o último elemento
    cleaned_row = row[:-1]  # Remove o último elemento
    cleaned_samples.append(cleaned_row)

print("\nLabels:", labels_list)
print("\nAmostras sem o último elemento:", cleaned_samples)
print("\nNúmero de amostras únicas:", len(cleaned_samples))
```

## Reorganizar a ordem dos gases para h2, ch2, c2h2, c2h4, c2h6

```python
new_order = [0,1,4,3,2]

reorganized_data = [tuple(tupla[i] for i in new_order) for tupla in cleaned_samples]

print(reorganized_data)
```

## Escrever os arquivos csv

```python
df = pd.DataFrame(reorganized_data)
df.columns = ['h2','ch2','c2h2','c2h4','c2h6']
df.to_csv('Datasets/Ibrahim-444-dataset/test_samples.csv', index=False)

df = pd.DataFrame(labels_list)
df.columns = ['act']
df.to_csv('Datasets/Ibrahim-444-dataset/test_labels.csv', index=False)

df = pd.DataFrame(duplicates)
df.columns = ['h2','ch2','c2h2','c2h4','c2h6', 'act']
df.to_csv('Datasets/Ibrahim-444-dataset/duplicates.csv', index=False)
```
