```python
import csv
labels_file = "1.0-dataset/train_labels.csv"
samples_file = "1.0-dataset/train_samples.csv"
```

## LÃª o arquivo dos gases e os coloca em uma lista:

```python
def list_of_all_gases(filename = samples_file):
    gases = []
    with open(filename, mode='r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)
        for row in csv_reader:
            gases.append(row)
    return gases

list_of_gases = list_of_all_gases()
```


## Passa pela tabela DGA e armazena os resultados em uma lista:


```python
dga_table_labels = []
i = 0

def is_zero(gas):
    if gas is None or gas == 0.0: 
        return 0.001  
    return gas

while(i < len(list_of_gases)):
    h2 = float(list_of_gases[i][0]) 
    ch4 = float(list_of_gases[i][1])
    c2h2 = float(list_of_gases[i][2])
    c2h4 = float(list_of_gases[i][3]) 
    c2h6 = float(list_of_gases[i][4])
    
    h2 = is_zero(h2) 
    ch4 = is_zero(ch4)
    c2h2 = is_zero(c2h2)
    c2h4 = is_zero(c2h4) 
    c2h6 = is_zero(c2h6) 

    if (ch4 / h2 < 0.1) and (c2h4 / c2h6 < 0.2):
        dga_table_labels.append(f"{2}")  # PD
            
    elif (c2h2 / c2h4 > 1) and (0.1 <= ch4 / h2 <= 0.5) and (c2h4 / c2h6 > 1):
          dga_table_labels.append(f"{3}")  # D1

    elif (0.6 <= c2h2 / c2h4 <= 2.5) and (0.1 <= ch4 / h2 <= 1) and (c2h4 / c2h6 > 2):
          dga_table_labels.append(f"{4}")  # D2

    elif (ch4 / h2 > 1) and (c2h4 / c2h6 < 1):
          dga_table_labels.append(f"{5}")  # T1

    elif (c2h2 / c2h4 < 0.1) and (ch4 / h2 > 1) and (1 <= c2h4 / c2h6 <= 4):
          dga_table_labels.append(f"{6}")  # T2

    elif (c2h2 / c2h4 < 0.2) and (ch4 / h2 > 1) and (c2h4 / c2h6 > 4):
          dga_table_labels.append(f"{7}")  # T3
    
    else: 
        dga_table_labels.append(f"{1}") # NF
    i += 1

print(len(dga_table_labels))
```

## Imprime as labels reais e esperadas lado a lado:

```python

gases = []
equal_counter = 0
with open(labels_file, mode='r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    i = 0
    for row in csv_reader:
        print(f"Real: {row[0]}  |   Esperado: {dga_table_labels[i]} ")
        if float(row[0]) == float(dga_table_labels[i]):
            equal_counter += 1
        i+=1

print(f"{equal_counter} valores iguais")
```
